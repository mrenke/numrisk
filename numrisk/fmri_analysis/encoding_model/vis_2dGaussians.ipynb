```python
import numpy as np
import pandas as pd

```

```python
x = np.linspace(0, 10, 100)
y = x

import scipy.stats as stats
sd = 2
pdf_x = stats.norm.pdf(x, 7, sd)
pdf_y = stats.norm.pdf(y,3, sd)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, pdf_x, label='pdf_x')
```

```python
import seaborn as sns

data_2d = pdf_x + pdf_y[:, np.newaxis]
plt.figure(figsize=(4,4))
sns.heatmap(data_2d, ).set_title('sum of pdfs') # xticklabels=False, yticklabels=False

plt.figure(figsize=(4,4))
data_2d = pdf_x * pdf_y[:, np.newaxis]
sns.heatmap(data_2d, xticklabels=False, yticklabels=False).set_title('product of pdfs')
```

```python
plt.figure(figsize=(4,4))
data_2d = (pdf_x+pdf_y) * pdf_y[:, np.newaxis]
sns.heatmap(data_2d, xticklabels=False, yticklabels=False).set_title('product of pdfs')
```
