```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os.path as op
import os

bids_folder = '/Users/mrenke/data/ds-dnumrisk'
plot_folder = '/Users/mrenke/data/ds-dnumrisk/plots_and_ims/behavior_risk/' 
sns.set_context('talk')

subList_remove = [32,40,45,46,50]
```

```python
from utils import get_data

df = get_data()
df['group'] = np.where(df['group'] == 0, 'control', 'dyscalc')
df['x'] = df['log(risky/safe)'] 
```

```python
df.head()
```

```python
import bambi
format = 'symbolic'
sub = 1

data = df.xs(sub, level='subject')
data = data.xs(format, level='format')
model = bambi.Model('chose_risky ~ x + n_safe + n_risky', link='probit', family='bernoulli', data=data.reset_index())

traces = model.fit(init='adapt_diag', target_accept=0.9, draws=1000, tune=1000)
```

```python
import arviz as az
model_label = f'testHeuristicSub-{sub}'
target_folder = op.join(bids_folder, 'derivatives', 'cogmodels_risk')

az.to_netcdf(traces, op.join(target_folder, f'rnp_model-{model_label}_format-{format}_trace.netcdf'))
```

```python
variable_names = list(traces.posterior.data_vars.keys())

fig, axs = plt.subplots(1,len(variable_names), figsize=(10,2), sharex=False)

for i,var in enumerate(variable_names):
        temp = traces.posterior[var].to_dataframe()
        count_greater_zero = np.round(np.mean(temp[var] > 0),3)
        sns.kdeplot(temp, ax=axs[i],legend=False, fill=True)
        axs[i].axvline(x=0, color='r', linestyle='--')
        axs[i].set_title(f'{var},\n p<{count_greater_zero}')
        print(f'{var}: mean={temp[var].mean()}, 95%CI={np.percentile(temp[var], [2.5, 97.5])}')
        
plt.savefig(op.join(plot_folder, f'rnp_model-{model_label}_format-{format}_regressors.pdf'), bbox_inches='tight')
        
```

```python
for sub in subList_remove:
    model_label = f'testHeuristicSub-{sub}'

    data = df.xs(sub, level='subject')
    data = data.xs(format, level='format')
    model = bambi.Model('chose_risky ~ x + n_safe + n_risky', link='probit', family='bernoulli', data=data.reset_index())

    traces = model.fit(init='adapt_diag', target_accept=0.9, draws=1000, tune=1000)
    
    variable_names = list(traces.posterior.data_vars.keys())

    fig, axs = plt.subplots(1,len(variable_names), figsize=(12,2), sharex=False)

    for i,var in enumerate(variable_names):
            temp = traces.posterior[var].to_dataframe()
            count_greater_zero = np.round(np.mean(temp[var] > 0),3)
            sns.kdeplot(temp, ax=axs[i],legend=False, fill=True)
            axs[i].axvline(x=0, color='r', linestyle='--')
            axs[i].set_title(f'{var},\n p<{count_greater_zero}')
            print(f'{var}: mean={temp[var].mean()}, 95%CI={np.percentile(temp[var], [2.5, 97.5])}')

    plt.savefig(op.join(plot_folder, f'rnp_model-{model_label}_format-{format}_regressors.pdf'), bbox_inches='tight')
```
